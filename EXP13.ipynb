EXP-13 Create an infographic using Matplotlib summarizing the key insights from a dataset about employee performance ratings.
OUTPUT:  A combination of bar charts, pie charts, and annotations summarizing performance ratings.


1. Load the dataset.
2. Aggregate data: calculate rating distribution, average projects, and attendance.
3. Create subplots for each data point.
4. Plot performance rating distribution as a bar chart.
5. Plot average projects completed as a bar chart.
6. Plot average attendance as a bar chart.
7. Add an overall title.
8. Adjust layout and display the infographic.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset from CSV
df = pd.read_csv('employee_performance.csv')

# Step 2: Aggregate data for visualization
rating_counts = df['Performance Rating'].value_counts()
average_projects = df.groupby('Performance Rating')['Projects Completed'].mean()
average_attendance = df.groupby('Performance Rating')['Attendance (%)'].mean()

# Step 3: Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Bar Chart: Distribution of Performance Ratings
axs[0].bar(rating_counts.index, rating_counts.values, color=["green", "blue", "orange"])
axs[0].set_title("Performance Rating Distribution", fontsize=14)
axs[0].set_ylabel("Number of Employees", fontsize=12)
axs[0].set_xlabel("Performance Rating", fontsize=12)
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Bar Chart: Average Projects Completed by Rating
axs[1].bar(average_projects.index, average_projects.values, color=["green", "blue", "orange"])
axs[1].set_title("Average Projects Completed", fontsize=14)
axs[1].set_ylabel("Projects", fontsize=12)
axs[1].set_xlabel("Performance Rating", fontsize=12)
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

# Bar Chart: Average Attendance by Rating
axs[2].bar(average_attendance.index, average_attendance.values, color=["green", "blue", "orange"])
axs[2].set_title("Average Attendance (%)", fontsize=14)
axs[2].set_ylabel("Attendance (%)", fontsize=12)
axs[2].set_xlabel("Performance Rating", fontsize=12)
axs[2].grid(axis='y', linestyle='--', alpha=0.7)

# Add an overall title
plt.suptitle("Employee Performance Summary", fontsize=16)

# Adjust layout and display the infographic
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
