EXP-06 : Create a scatterplot to analyze the relationship between advertising expenses and sales from a dataset. 
OUTPUT :  A scatterplot showing the positive or negative correlation between expenses and sales.

Step-1: Import pandas for data handling and matplotlib.pyplot for plotting.
Step-2: Load the dataset using pd.read_csv() to create a DataFrame.
Step-3: Inspect the dataset by displaying the first few rows with df.head().
Step-4: Create scatterplots for each advertising type (TV, Radio, Newspaper) against sales using plt.scatter(), applying different colors for each type.
Step-5: Set the plot title with plt.title() and label the x-axis and y-axis using plt.xlabel() and plt.ylabel().
Step-6: Add a legend with plt.legend() to distinguish between the different advertising types.
Step-7: Use plt.show() to display the scatterplot.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('advertising.csv')

# Display the first few rows of the dataset to check the structure
print(df.head())

# Create a scatterplot to visualize the relationship between advertising expenses and sales
plt.figure(figsize=(8,6))
plt.scatter(df['TV'], df['Sales'], color='blue', label='TV Advertising', alpha=0.5)
plt.scatter(df['Radio'], df['Sales'], color='red', label='Radio Advertising', alpha=0.5)
plt.scatter(df['Newspaper'], df['Sales'], color='green', label='Newspaper Advertising', alpha=0.5)

# Add titles and labels
plt.title('Scatterplot of Advertising Expenses vs Sales')
plt.xlabel('Advertising Expenses ($)')
plt.ylabel('Sales ($)')
plt.legend()

# Show the plot
plt.show()
```
