EXP-07: Use pandas to group aggregate a dataset on city wise covid -19 cases . visualize the data with a bar chart .
OUTPUT : Abar chart showing total cases for each city.

1. Import necessary libraries: `pandas` for data manipulation, `matplotlib.pyplot` for plotting, and `seaborn` for enhanced visualization.
2. Load the dataset using `pd.read_csv()` to create a DataFrame.
3. Group data by 'City' and sum 'Total_Cases' to aggregate the total cases for each city.
4. Sort the data in descending order by 'Total_Cases' for better visualization.
5. Plot a bar chart using `sns.barplot()`, add labels, and display the plot with `plt.show()`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('covid_data.csv')  # Make sure the file is in the same directory or provide the full path

# Grouping data by 'City' and summing the 'Total_Cases' for each city (in case there are multiple rows per city)
city_cases = df.groupby('City')['Total_Cases'].sum().reset_index()

# Sorting the data by Total_Cases for better visualization
city_cases = city_cases.sort_values(by='Total_Cases', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='Total_Cases', y='City', data=city_cases, palette='coolwarm')

# Adding labels and title
plt.title('COVID-19 Total Cases by Indian City', fontsize=16)
plt.xlabel('Total Cases', fontsize=12)
plt.ylabel('City', fontsize=12)

# Display the plot
plt.show()
```
