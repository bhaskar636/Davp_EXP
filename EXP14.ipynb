 EXP-14: Build a real-time updating graph using Dash to visualize live weather conditions like temperature, humidity, and wind speed.
OUTPUT : A live-updating dashboard showing line graphs for temperature, humidity, and wind speed.

1. Load the dataset using `pd.read_csv()`.
2. Convert the 'Formatted Date' column to datetime format using `pd.to_datetime()`.
3. Create a plot with three lines for temperature, humidity, and wind speed.
4. Set the plot title, x-axis, and y-axis labels.
5. Rotate the x-axis labels for better readability.
6. Display the legend and grid.
7. Adjust the layout to ensure everything fits.
8. Show the plot using `plt.show()`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('weather_data14.csv')  # Replace with the correct file name

# Display the dataset
print("Dataset Preview:")
print(df)

# Ensure 'Formatted Date' is in datetime format
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])

# Plot the data
plt.figure(figsize=(10, 6))

plt.plot(df['Formatted Date'], df['Temperature (C)'], label='Temperature (C)', color='red', marker='o')
plt.plot(df['Formatted Date'], df['Humidity'], label='Humidity', color='blue', marker='x')
plt.plot(df['Formatted Date'], df['Wind Speed (km/h)'], label='Wind Speed (km/h)', color='green', marker='s')

# Add title and labels
plt.title('Weather Trends (Reduced Dataset)', fontsize=16)
plt.xlabel('Date and Time', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
