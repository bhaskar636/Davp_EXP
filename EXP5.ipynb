EXP-5 : Perform exploratory data analysis (EDA) on a dataset containing student scores in different subjects. Show correlations using a heatmap.

OUTPUT:  A heatmap visualizing correlations between subjects like Math, Science, and English.

Step-1: Import Libraries: Import pandas, seaborn, and matplotlib for data handling and visualization.
Step-2: Load Data: Use pd.read_csv() to load the student dataset.
Step-3: Clean Data: Drop non-numeric columns (e.g., "Unnamed: 0", "Gender", "DOB").
Step-4: Correlation Matrix: Use .corr() to calculate correlations between subjects.
Step-5:  Plot the correlation matrix using sns.heatmap() and display with plt.show().

```python
# Step 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
# Replace the file path with your actual file path if needed
df = pd.read_csv('student_marks.csv')

# Step 3: Inspect the dataset
# Drop the non-numeric columns ('Unnamed: 0', 'Gender', 'DOB') to focus on the numerical data
df_numeric = df.drop(columns=['Unnamed: 0', 'Gender', 'DOB'])

# Step 4: Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Step 5: Create a heatmap to visualize correlations
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
plt.title("Correlation Heatmap of Student Scores in Different Subjects")
plt.show()
```
