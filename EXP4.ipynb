EXP-04: Use web scraping to extract weather data from a website. Visualize the temperature trends over a week using a line chart.
OUTPUT: A line chart showing temperature variations across 7 days.


1. Load the dataset using `pd.read_csv()`.
2. Verify that the dataset contains 'Formatted Date' and 'Temperature (C)' columns.
3. Convert the 'Formatted Date' column to `datetime` format.
4. Filter the last 7 days of data using `df.tail(7)`.
5. Extract the 'Formatted Date' and 'Temperature (C)' columns for plotting.
6. Plot the temperature trend using `matplotlib`, with dates on the x-axis and temperatures on the y-axis.
7. Display the plot with labels, title, grid, and legend.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('weatherHistory.csv')  # Replace with your actual dataset path

# Step 2: Check the first few rows of the dataset to verify columns
print(df.head())

# Step 3: Ensure the dataset has 'Formatted Date' and 'Temperature (C)' columns
if 'Formatted Date' not in df.columns or 'Temperature (C)' not in df.columns:
    raise ValueError("Dataset must have 'Formatted Date' and 'Temperature (C)' columns")

# Step 4: Convert 'Formatted Date' to datetime format (in case it's not already)
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])

# Step 5: Filter the data for the past week (or any 7-day period)
df_last_week = df.tail(7)  # Taking the last 7 entries, modify if needed for different time range
# Step 6: Extract the relevant data (formatted date and temperature)
dates = df_last_week['Formatted Date']
temperature_data = df_last_week['Temperature (C)']

# Step 7: Visualize the temperature trend over a week
plt.figure(figsize=(10, 6))
plt.plot(dates, temperature_data, marker='o', color='b', label='Temperature (°C)')
plt.title('Temperature Trends Over the Last Week', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)
plt.legend()
plt.tight_layout()

# Step 8: Show the plot
plt.show()
```
